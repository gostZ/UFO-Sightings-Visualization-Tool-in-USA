```python
import pandas as pd
```

```python
df = pd.read_csv('ufo.csv',index_col = None)
df = pd.DataFrame(df)
df = df[df['country'] == 'us']
df.dropna(subset=['encounter_length'],inplace=True)
df.dropna(subset=['state'],inplace=True)
df.dropna(subset=['date_documented'],inplace=True)
df.dropna(subset=['latitude'],inplace=True)
df.dropna(subset=['longitude'],inplace=True)
df.dropna(subset=['ufo_shape'],inplace=True)
df.dropna(subset=['date_time'],inplace=True)
df = df.drop(df[df['state'] == ''].index)
df = df.drop(df[df['ufo_shape'].isin(['changed', 'changing','NA','other','unknown'])].index)
```

```python
df[['Date','Time']] = df['date_time'].str.split(' ', expand=True)
df[['Month','Day','Year']] = df['Date'].str.split('/', expand=True)
```

```python
df = df.drop(['country', 'described_encounter_length','description','date_documented','Date','Year','Day','Time'], axis=1)
# df = df.drop([11627, 2])
df
```
